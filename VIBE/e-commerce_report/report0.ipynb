```python
import pandas as pd
import os

all_df = []
error_file = []

# Desired column headers
desired_columns = [
    'Franchise \nStore', 'Business\nDate', 'Date Time\nPlaced',
    'Date Time\nFulfilled', 'Net\nAmount', 'Quantity', 'Royalty\nItem',
    'Taxable\nItem', 'Order Id', 'Item Id', 'Menu Item\nName',
    'Menu Item\nAccount', 'Bundle Name', 'Employee',
    'Override\nApproval Employee', 'Order\nPlaced\nMethod',
    'Order\nFulfilled\nMethod', 'Modified\nOrder\nAmount',
    'Modification Reason', 'Payment\nMethods', 'Refunded',
    'Tax\nIncluded\nAmount'
]
folder_path = '/Users/nooreen/Desktop/High_margin/VIBE/LC/lc'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
# List all CSV files in current directory (assuming Google Colab after upload)
# csv_files = [file for file in os.listdir() if file.endswith('.csv')]

for file_name in csv_files:
    path=os.path.join(folder_path, file_name)

    try:
        with open(path, 'r', encoding='latin') as f:
            lines = f.readlines()

        # Find the line index that matches desired_columns
        header_line_index = None
        for idx, line in enumerate(lines):
            cols = [col.strip() for col in line.strip().split(',')]
            if cols == desired_columns:
                header_line_index = idx
                break

        if header_line_index is not None:
            df = pd.read_csv(path, encoding='latin', skiprows=header_line_index)
            all_df.append(df)
        else:
            raise ValueError("Desired header not found.")

    except Exception as e:
        print(f"Error in smart skip for {file_name}: {e}")
        error_file.append(file_name)
```

```python
import pandas as pd
import plotly.graph_objects as go

# 1. Aggregate revenue at the lowest level
agg = data.groupby(['Brand', 'channel', 'OrderMode'], as_index=False)['revenue'].sum()

# 2. Create label list
big_brand = ['VIBE']
brands = agg['Brand'].unique().tolist()
channels = agg['channel'].unique().tolist()
modes = agg['OrderMode'].unique().tolist()
labels = big_brand + brands + channels + modes

# 3. Create label index map
label_indices = {label: i for i, label in enumerate(labels)}

# 4. Build links
source = []
target = []
value = []
custom_labels = []

# VIBE → Brand
for brand in brands:
    brand_total = agg[agg['Brand'] == brand]['revenue'].sum()
    source.append(label_indices['VIBE'])
    target.append(label_indices[brand])
    value.append(brand_total)
    custom_labels.append(f"VIBE → {brand}<br>Revenue: ${brand_total:,.2f}")

# Brand → Channel
for _, row in agg.groupby(['Brand', 'channel'])['revenue'].sum().reset_index().iterrows():
    source.append(label_indices[row['Brand']])
    target.append(label_indices[row['channel']])
    value.append(row['revenue'])
    custom_labels.append(f"{row['Brand']} → {row['channel']}<br>Revenue: ${row['revenue']:,.2f}")

# Channel → OrderMode
for _, row in agg.iterrows():
    source.append(label_indices[row['channel']])
    target.append(label_indices[row['OrderMode']])
    value.append(row['revenue'])
    custom_labels.append(f"{row['channel']} → {row['OrderMode']}<br>Revenue: ${row['revenue']:,.2f}")

# 5. Plot Sankey diagram with hover labels
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=source,
        target=target,
        value=value,
        customdata=custom_labels,
        hovertemplate='%{customdata}<extra></extra>'
    )
))

fig.update_layout(title_text="Revenue Flow: VIBE → Brand → Channel → Order Mode", font_size=12)
fig.show()
```
