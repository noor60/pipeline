```python
item_to_consider=[
'Stuffed Crazy Crust Pepperoni',
'Stuffed Crazy Crust Cheese',

'Pepperoni Crazy Puffs®',
'Cheese and Herb Crazy Puffs®',
'Bacon & Cheese Crazy Puffs®',
'4 Cheese Crazy Puffs®',
'Ultimate Supreme',
'3 Meat Treat®',
'Hula Hawaiian®',
'Veggie',
'Detroit-Style Deep Dish Pepperoni',
'Detroit-Style Deep Dish Cheese',
'Detroit-Style Deep Dish Sausage',
'Caesar Wings® – Oven Roasted'	]


len(item_to_consider)
```

```python
print(scraped_data[scraped_data['item_name'].isin(item_to_consider)]['item_name'].nunique())
processed_data=scraped_data[scraped_data['item_name'].isin(item_to_consider)]

# NOTE need ceaser wings to manually pick price and 'Bacon & Cheese Crazy Puffs','Cheese and Herb Crazy Puffs', not present in menu.
```

```python
import numpy as np
processed_data['item_name']=np.where(processed_data['item_name']=='Caesar Wings® – Oven Roasted','Caesar Wings',processed_data['item_name'])
```

```python
processed_data['price']=np.where(processed_data['item_name']=='Caesar Wings',8.99,processed_data['price'])
processed_data[processed_data['item_name']=='Caesar Wings']
```

```python
processed_data['price']=np.where((processed_data['item_name']=='Caesar Wings')& (processed_data['store_id']==8374),9.99,processed_data['price'])
processed_data[processed_data['item_name']=='Caesar Wings']
```
