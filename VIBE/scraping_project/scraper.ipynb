```python
import requests
import pandas as pd

# Parameters
store_id =[ '10332','10750']
menu_version = '1'
base_url = 'https://api.cloud.littlecaesars.com/bff/api/v4/stores'

item_name=[]
item_price=[]
category=[]
store_ids=[]

for store in store_id:
    # Make the request
    url = f"{base_url}/{store}/{menu_version}/menu"
    response = requests.get(url)

    # Handle the response
    if response.status_code == 200:
        print("Menu data retrieved successfully.")
        menu_data = response.json()
        categories=menu_data['menuTypes']



        for categoryi in categories:
            cat=categoryi['typeDescription']
            # print(cat)
            

            menu_items=categoryi['menuItems'] #list of items

            for item in menu_items:
                name=item['itemName']
                price=item['price']

                item_name.append(name)
                item_price.append(price)
                store_ids.append(store)
                category.append(cat)
            
    else:
        print(f"Request failed with status code {response.status_code}")
        print(response.text)

        
df=pd.DataFrame({'item_name':item_name,'price':item_price,'category':category,'store_id':store_ids})
df
     
```
